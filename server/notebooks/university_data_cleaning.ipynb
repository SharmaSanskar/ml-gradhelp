```python
import numpy as np
import pandas as pd
```

```python
colnames = ['Names', 'Tuition', 'City', 'State']
```

```python
df1 = pd.read_csv('../datasets/USNewsRankings.csv', usecols=[1, 3, 5, 6], names=colnames, header=None)
```

```python
df1['Rank'] = df1.index + 1
```

```python
df1.head()
```

```python
df2 = pd.read_csv('../datasets/Niche.csv')
df2.head()
```

```python
merge_df = pd.merge(df1, df2, on='Names')
```

```python
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df['Location'] = merge_df['City'] + ", " + merge_df['State']
```

```python
unwanted_cols = ['City', 'State', 'Niche Link', 'Atheletic Division', 'Undergrad #', 'Graduation Rate']

final_df = merge_df.drop(unwanted_cols, axis=1)
```

```python
final_df.columns = ['name', 'tuition', 'rank', 'acceptance', 'website', 'location']
```

```python
final_df.head()
```

```python
final_df.to_csv('../datasets/universities_data_300.csv', index=False)
```

```python
final_df.to_json('../datasets/universities.json', orient='records')
```
